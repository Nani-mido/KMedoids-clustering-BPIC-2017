```python
#slicing DataFrame by case id to take one line only from each case
grouping=data.groupby(['Case ID']).apply(lambda x:(x.tail(1)))

#counting cases in each variant
v=grouping['Variant index'].unique()
ls=[]
for i in list(v):
    ls.append(len(grouping[grouping['Variant index']==i]))

#make it as a data frame to easy manipulating
variants=pd.DataFrame(columns=['variant','count'])
variants['variant']=grouping['Variant index'].unique()
variants['count']=ls

# sorting by most common variant
variants=variants.sort_values(by=['count'],ascending=False)

#taking the most common variant (in which there is more than one case : first 1111 rows)
#data1=data[data['Variant index'].isin(list(variants['variant'][0:46]))]

#then taking the rest !!
#data2=data[data['Variant index'].isin(list(variants['variant'][46:]))]
```
